```python
#modified from https://blog.csdn.net/a486259/article/details/84787288
import cv2
import os
sourceFileName=''
video_path = os.path.join("E:", sourceFileName+'.mp4')
times=0
#The frequency of video extraction, one for every frameFrequency frame
frameFrequency=124
#Output pictures to the current directory vedio folder
outPutDirName='vedio/'+sourceFileName+'/'
if not os.path.exists(outPutDirName):
    #If the file directory does not exist, create a directory
    os.makedirs(outPutDirName) 
camera = cv2.VideoCapture(video_path)
while True:
    times+=1
    res, image = camera.read()
    if not res:
        print('not res , not image')
        break
    if times%frameFrequency==0:
        cv2.imwrite(outPutDirName + str(times)+'.jpg', image)
        print(outPutDirName + str(times)+'.jpg')
print('Picture extraction is over')
camera.release()
```
